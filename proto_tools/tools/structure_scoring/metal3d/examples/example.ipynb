# Metal3D Prediction

This example scores the 2VVB carbonic anhydrase zinc site with the default original Metal3D checkpoint and reports per-residue probabilities for the His94/His96/His119 zinc-coordinating triad. Switch `model_checkpoint` to `metal3d-cat` or `metal3d-clean` to use dEVA's retrained variants.

```python
from proto_tools import (
    Metal3DPredictionConfig,
    Metal3DPredictionInput,
    Metal3DStructureInput,
    Structure,
    run_metal3d_prediction,
)

structure = Structure(
    structure="https://raw.githubusercontent.com/gelnesr/dEVA/main/inputs/2VVB.pdb"
)

output = run_metal3d_prediction(
    inputs=Metal3DPredictionInput(
        inputs=[
            Metal3DStructureInput(
                structure=structure,
                candidate_residues={"X": [94, 96, 119]},
            )
        ]
    ),
    config=Metal3DPredictionConfig(
        model_checkpoint="metal3d-original",
        probability_threshold=0.2,
        device="cuda",
    ),
)

output.results[0].model_dump(mode="json")
```
